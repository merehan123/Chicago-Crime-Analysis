```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("chicago_crime_cleand.csv")
```

```python
df
```

```python
top_crimes = df["Primary Type"].value_counts().head(10)
sns.barplot(x=top_crimes.values, y=top_crimes.index)
plt.title("top 10 crime types")
plt.xlabel("count")
plt.ylabel("crime type")
plt.show()
```

```python
year_counts = df["Year"].value_counts().sort_index()
sns.lineplot(x=year_counts.index, y=year_counts.values)
plt.title("crimes per year")
plt.xlabel("year")
plt.ylabel("number of crimes")
plt.show()
```

```python
month_counts = df["Month"].value_counts().sort_index()
sns.lineplot(x=month_counts.index, y=month_counts.values)
plt.title("crimes by month")
plt.xlabel("month")
plt.ylabel("count")
plt.show()
```

```python
hour_counts = df["Hour"].value_counts().sort_index()
sns.lineplot(x=hour_counts.index, y=hour_counts.values)
plt.title("crimes by hour")
plt.xlabel("hour")
plt.ylabel("count")
plt.show()
```

```python
sns.countplot(x="Timeofday",data=df)
plt.title("crimes by time of day")
plt.show()
```

```python
sns.countplot(x="Dayofweek", data=df)
plt.title("crimes by day of week")
plt.show()
```

```python
sns.countplot(x="IsWeekend", data=df)
plt.title("weekend vs weekday crimes")
plt.show()
```

```python
top_locations = df["Location Description"].value_counts().head(10)
sns.barplot(x=top_locations.values, y=top_locations.index)
plt.title("top crime locations")
plt.show()
```

```python
sns.countplot(x="Arrest", data=df)
plt.title("arrest analysis")
plt.show()
```

```python
sns.countplot(x="Domestic", data=df)
plt.title("domestic crimes")
plt.show()
```

```python
sns.scatterplot(x=df["Longitude"],y=df["Latitude"],alpha=0.1)
plt.title("crime locations")
plt.show()
```
