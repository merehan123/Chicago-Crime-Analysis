```python
import pandas as pd
df = pd.read_csv("Chicago_Crimes_2012_to_2017.csv")
df.shape
df.head()
df.info()
df.describe()
```

```python
df.head()
```

```python
columns_drop = [
    "Unnamed: 0",
    "ID",
    "Case Number",
    "FBI Code",
    "Updated On",
    "Location"

]
df = df.drop(columns=columns_drop)
```

```python
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.dropna(subset=["Latitude", "Longitude"])
df.head()
```

```python
df = df.drop(columns=["X Coordinate", "Y Coordinate"])
```

```python
df["Location Description"].fillna("unknown", inplace=True)
df.head()
```

```python
df["Community Area"].fillna(df["Community Area"].mode()[0], inplace=True)
```

```python
df["Ward"].fillna(df["Ward"].mode()[0], inplace=True)
```

```python
df["District"].fillna(df["District"].mode()[0], inplace=True)
```

```python
df["District"] = df["District"].astype(int)
df["Ward"] = df["Ward"].astype(int)
df["Community Area"] = df["Community Area"].astype(int)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
import pandas as pd 
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Month"] = df["Date"].dt.month
```

```python
df["Day"] = df["Date"].dt.day
```

```python
df["Hour"] = df["Date"].dt.hour
```

```python
df["Dayofweek"] = df["Date"].dt.day_of_week
```

```python
df.head()
```

```python
df["IsWeekend"] = df["Dayofweek"].apply(lambda x: 1 if x >= 5 else 0)
```

```python
def Gettimeofday(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"
df["Timeofday"] = df["Hour"].apply(Gettimeofday)
```

```python
df.head()
```

```python
df = df.drop(columns=["Date"])
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates(subset=[
    "Day","Month","Year", "Primary Type", "Latitude", "Longitude"
])
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["Block"])
```

```python
df = df.drop(columns=["IUCR"])
```

```python
df["Year"].unique()
```

```python
df["Primary Type"] = df["Primary Type"].str.lower().str.strip()
df["Description"] = df["Description"].str.lower().str.strip()
df["Location Description"] = df["Location Description"].str.lower().str.strip()
df["Timeofday"] = df["Timeofday"].str.lower().str.strip()
```

```python
df.head()
```

```python
df.to_csv('chicago_crime_cleand.csv', index=False)
```
